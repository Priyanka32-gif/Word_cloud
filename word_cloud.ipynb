```python
import nltk
```

```python
text = data['Preamble'][0]
```

```python
nltk.download('stopwords')

from nltk.corpus import stopwords

stopwords = stopwords.words('english')
```

```python
import string
```

```python
string.punctuation
```

```python
text
```

```python
nopunc = [c for c in  text if c not in string.punctuation]
```

```python
nopunc
```

```python
stopwords
```

```python
nopunc = ''.join(nopunc)
```

```python
nopunc
```

```python
nopunc.split()
```

```python
clean_text = [word for word in nopunc.split() if word.lower() not in stopwords]
```

```python
clean_text
```

```python
clean_text = " ".join(clean_text)
```

```python
clean_text
```

```python
wordcloud = WordCloud(background_color='white',max_font_size=50,max_words=100).generate(clean_text)
```

```python
plt.imshow(wordcloud)
```
